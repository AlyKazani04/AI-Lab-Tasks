```python
# 3

num = int(input("Enter number of numbers to enter: "))

li = []
for i in range(num):
    li.append(int(input(f"Enter element {i}: ")))

print(li)
sum = 0
min = li[0]
max = li[0]
for i in li:
    sum += i
    if i <= min:
        min = i
    if i >= max:
        max = i
print(f"Sum: {sum}")
print(f"Min: {min}")
print(f"Max: {max}")
index = int(input(f"Enter index to update(0 - {num - 1}): "))
val = int(input(f"Enter updated value: "))
li[index] = val
print(li)
```

```python
# 4

def waive_discount(amount):
    if amount < 1000:
        discount = .05
    elif amount < 5000:
        discount = .1
    else:
        discount = 0.15
    return amount - amount * discount

amount = float(input("Enter purchase amount: "))
print(f"Payable Amount: {waive_discount(amount)}")
```

```python
# 5

def create_dict(text : str):
    temp = {
        "characters" : len(text),
    }
    
    vowels, consonants = 0,0
    
    for c in text:
        if c in ('a', 'e', 'i', 'o', 'u'):
            vowels += 1
        elif c == ' ': 
            continue
        else:
            consonants += 1
    temp.update({ 
        "vowels" : vowels,
        "consonants" : consonants
    })
    return temp

val = str(input("Enter a string: "))
# input: hello world this is aly
new_dict = dict(create_dict(val))
print(new_dict)
```

```python
# 6

text = str(input("Enter a string: "))
# input: hello world

substrs = []
for i in range(len(text)):
    substr = ""
    for j in range(i, len(text)):
        if text[j] in substr:
            substrs.append(substr)
            substr = ""
            break
        else:
            substr += text[j]

    substrs.append(substr)

longest = substrs[0]
for subs in substrs:
    if len(subs) > len(longest):
        longest = subs

print(f"Longest substring: '{longest}'")
```

```python
# 7

def give_repeated(li : list):
    num_set = set()
    repeated = []
    for i in li:
        if i in num_set:
            if i not in repeated:
                repeated.append(i)
        num_set.add(i)
    return sorted(repeated)

li = [1,2,4,2,5,6,7,4,4,5,8,9]

print(f"Repeated letters in {li}:\n{give_repeated(li)}")
```
